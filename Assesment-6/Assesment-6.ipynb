```python
## 1. Write a Python program that can do the following:
#- You have $50
#- You buy an item that is $15, that has a 3% tax
#- Print how much money you have left, after purchasing the item.

amount = 50
buy_item = 15
tax = buy_item * (3/100)
money_left = amount -(buy_item + tax)
print(f"Money Left :{money_left}$")
```

```python
# 2. Ask the user how many days until their birthday
#- Print an approx. number of weeks until their birthday
#- 1 week is = to 7 days.

bdy_days_left = int(input("Enter how many days left until your birthday :"))
week_left = bdy_days_left / 7
print("Week Left :",week_left)
```

```python
#3. Create a variable “Grade” holding an integer between 0 to 100. Code to print
#the letter grade of the number grade variable.
#Grades:
#- A = 90 - 100
#- B = 80 - 89
#- C = 70-79
#- D = 60 - 69
#- F = 0 – 59

Grades = int(input("Enter marks between 0 to 100:"))
if 0 <= Grades <= 100:
    if 90 <= Grades <=100:
        print("A")
    elif 80 <= Grades <=89:
        print("B")
    elif 70 <= Grades <=79:
        print("C")
    elif 60 <= Grades <=69:
        print("D")
    else:
        print("F")
else:
    print("Invalid input")
```
