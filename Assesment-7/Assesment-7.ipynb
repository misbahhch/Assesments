```python
# 3. Write a program to calculate the sum of numbers from 1 to N entered by the user.
N = int(input("Enter num :"))
sum = 0
for i in range(1,N+1):
    sum = sum + i 
print("Sum: ", sum)
    
    
    
        
```

```python
# 4.Print the multiplication table of a number entered by the user (up to 10).
multi_num = int(input("Enter num :"))
result =  0
for i in range(1, 11):
    result = multi_num * i 
    print(f'{multi_num} * {i} = {result}')
    
```

```python
# 5.Write a program to reverse a string using a for loop.
string = input("Enter string :")
reverse = ""
for i in string:
    reverse = i + reverse
print(reverse)
```

```python
# 6. Count the number of vowels present in a string using a for loop.
vowels = ['a', 'e' ,'i' ,'o','u','A','E','I','O','U']
str1 = input("Enter string :")
count = 0
for i in str1:
    for j in vowels:
        if i == j:
            count = count + 1
print(count)
        
```

```python
# 8.Write a program to check whether a number is prime using a for loop.
num = int(input("Enter num:"))
is_prime = True
if num > 1:
    for i in range(2,num):
        if num % i == 0 :
          is_prime = False
          break
    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
         
```

```python
# 9.Given a list:
# numbers = [12, 5, 7, 18, 21, 30]
# Print only numbers divisible by 3.

numbers = [12, 5, 7, 18, 21, 30]
for i in numbers:
    if i % 3 == 0:
        print(i)
        
```

```python
# 10. Break and Continue
# Print numbers from 1 to 20 but: - Skip multiples of 3. - Stop execution if number becomes greater than 15.

for i in range(1 ,21):
    if i % 3 == 0:
        continue
    elif i > 15 :
        break
    print(i)
```

```python
# 12.Calculate factorial of a number using a while loop.
num = int(input("Enter number :"))
factorial = 1
i = 1
while i <= num:
    factorial = factorial * i
    i = i+1
print(f"Factorial of {num} is {factorial}")
```
