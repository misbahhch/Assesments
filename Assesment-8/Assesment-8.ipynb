```python
# 1. Student Grade Calculator
#Write a function calculate_grade(marks) that takes marks as input and returns grade according to:

#90 and above → A
#80 – 89 → B
#70 – 79 → C
#Below 70 → Fail

def calculate_grade(marks):
    if 0<= marks <=100:
        if marks >= 90:
            return "Grade: A"
        elif marks >= 80:
            return "Grade: B"
        elif marks >= 70:
            return "Grade: C"
        else:
            return "Fail"
    else:
        return "Invalid input"

print(calculate_grade(70))  
```

```python
# 2. Even or Odd Number Checker
#Write a function is_even(number) that returns:

#True if the number is even
#False if the number is odd

def is_even(number):
    if number %2 ==0:
        return True
    else:
        return False
print(is_even(44))
print(is_even(67))
```

```python
# 3. Factorial Using Function
#Write a function factorial(n) that returns the factorial of a number using a loop.
def factorial(n):
    fact = 1
    for i in range(1, n +1):
        fact = fact * i
            
    return fact
print(f'Factorial :{factorial(9)}')
            
```

```python
# 4. Find Maximum Number
# Write a function find_max(a, b, c) that returns the largest number among three numbers.
def find_max(a,b,c):
    if a >= b and a >= c:
        return f"Largest number :{a}"
    elif b >= a and b >= c:
        return f"Largest number :{b}"
    else:
        return c
print(find_max(4,6,5))
```

```python
# 5. Word Counter
# Write a function count_words(sentence) that counts how many words exist in a given sentence.
# Example:
# "I love Python programming"
# Output → 4

def count_words(sentence):
    words = sentence.split()
    return len(words)
count_words("I love Python Programming")
```

```python
# 6. Prime Number Checker
# Write a function is_prime(n) that checks whether a number is a prime number.
def is_prime(n):
    if n <= 1:
        return "Not a Prime Number"
    for i in range(2,n):
        if n % i == 0:
            return "Not a Prime Number"
        
    return "Prime number"
is_prime(11)
```

```python
# 7. List Sum Function
#Write a function list_sum(numbers) that takes a list as input and returns the sum of all numbers.

#Example:
#[10, 20, 30]
#Output → 60
def list_sum(numbers):
    sum = 0
    for i in numbers:
        sum = sum + i
    return sum
list_sum([10,20,30])
        
```

```python
# 8. Password Strength Checker
# Write a function check_password(password) that checks:

# Minimum 8 characters required
# Must contain at least one digit
# Return:

# "Strong" or "Weak"

def check_password(password):
    if len(password)<8:
        return "Weak"
    has_digit = False
    for i in password:
        
        if i >='0' and i<='9':
            has_digit = True
            break
    if has_digit == True:
        return "Strong"
    else:
        return "Weak"
            
print(check_password("Zoha12@#"))
print(check_password("Zohaaa"))
print(check_password("Zoh2345"))
            
```

```python
# 9. Temperature Converter
# Write a function celsius_to_fahrenheit(c) that converts temperature from Celsius to Fahrenheit.

# Formula:
# F = (C × 9/5) + 32
def celsius_to_fahrenheit(C):
    F = (C *9/5) + 32
    return f"Fahrenheit :{F}"
print(celsius_to_fahrenheit(37))
```

```python
# 10. Remove Duplicates from List
# Write a function remove_duplicates(data) that removes duplicate elements from a list and returns a new list.

# Example:
# [1,2,2,3,4,4]
#Output → [1,2,3,4]

def remove_duplicates(data):
    new_list = list(set(data))
    return new_list
remove_duplicates([1,2,2,4,6,6])
    
```
